```python
%%time
import time
import sys
import random
import numpy as np
import multiprocessing as mp
from pathlib import Path
from ont_fast5_api.fast5_interface import get_fast5_file
from scipy import stats

def process_fast5_read(read, window, skip=1000, zscore=True):
    """ Normalizes and extracts specified region from raw signal """

    s = read.get_raw_data(scale=True)  # Expensive

    if zscore:
        s = stats.zscore(s)

    pos = random.randint(skip, len(s)-window)

    return s[pos:pos+window].reshape((window, 1))

class MyMixedDatasetTrain(IterableDataset):
    def __init__(self, files, label, window):
        #TODO shuffle files
        self.files = files
        self.label = label
        self.window = window

    def __iter__(self):
        for fast5 in self.files:
            with get_fast5_file(fast5, mode='r') as f5:
                for read in f5.get_reads():
                    x = process_fast5_read(read, self.window)
                    y = np.array(self.label)
                    yield (x, y)
                    
dset = MyMixedDatasetTrain(fast5s_positives[:-3], 1, 1000)
train_loader = DataLoader(
    dset, 
    batch_size=128, 
    num_workers = 0, 
    # pin_memory=True,
    # persistent_workers=True,
)
for i in range(30):
    print(i)  
    next(iter(train_loader))
```

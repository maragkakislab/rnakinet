```python
from rnamodif.data_utils.dataloading2 import nanopore_datamodule
from rnamodif.data_utils.split_methods import get_kfold_split_func
from rnamodif.data_utils.split_methods import kfold_split


split_method=kfold_split(total_k=5, current_k=0)
s1 = split_method(pos_files='5eu_2022_nia', neg_files='UNM_2022_nia')
s2 = split_method(pos_files='5eu_2020_nia', neg_files='UNM_2020_nia')
dm = nanopore_datamodule(
    splits = [s1,s2],
    verbose=2, 
    batch_size=16, 
    valid_limit=1000, 
    workers=16,
    window=4096,
)
dm.setup()
```

```python
import torch
def check_loader(loader, batch_limit=None):
    total = 0
    positives = 0
    for i,(x,y) in enumerate(loader):
        if(batch_limit and i>=batch_limit):
            break
        total += len(y)
        positives += y.sum().int().numpy()
        assert len(torch.unique(x, dim=0)) == len(x)

    print('DONE')
    print('positives', positives,'/',total)

# check_loader(dm.val_dataloader(), batch_limit=10)
check_loader(dm.train_dataloader(), batch_limit=10)

```
